```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
y=train_data.Survived
```

```python
train_data=train_data.drop(['Survived'],axis='columns')
```

```python
data=pd.concat([train_data,test_data],axis='rows')
```

```python
data.shape
```

```python
418+891
```

```python
data.head()
```

```python
data.drop(['Cabin'],axis='columns',inplace=True)
```

```python
data.isnull().sum()
```

```python
data.Fare=data.Fare.fillna(data.Fare.median())
```

```python
data.Embarked=data.Embarked.fillna('S')
```

```python
data.Age.describe()
```

```python
data.Age=data.Age.fillna(data.Age.median())
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
data['Fam']=data['SibSp']+data['Parch']
```

```python
data=data[['PassengerId','Pclass','Sex','Age','Fare','Embarked','Fam']]
```

```python
data.set_index(['PassengerId'],inplace=True)
```

```python
data
```

```python
pclass_dummy=pd.get_dummies(data.Pclass,drop_first=True)
```

```python
pclass_dummy
```

```python
embarked_dummy
```

```python
sex_dummy=pd.get_dummies(data.Sex,drop_first=True)
embarked_dummy=pd.get_dummies(data.Embarked,drop_first=True)
```

```python
data=pd.concat([data,pclass_dummy,sex_dummy,embarked_dummy],axis='columns')
```

```python
data.drop(['Pclass','Sex','Embarked'],inplace=True,axis='columns')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
data=pd.DataFrame(data)
```
