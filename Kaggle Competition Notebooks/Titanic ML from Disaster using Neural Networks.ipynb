```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout
```

```python
def create_model(layers):
    model=Sequential()
    for i,nodes in enumerate(layers):
        if i==0:
            model.add(Dense(units=nodes,activation='relu',kernel_initializer='he_uniform',input_dim=8))
            model.add(Dropout(0.2))
        else:
            model.add(Dense(units=nodes,activation='relu',kernel_initializer='he_uniform'))
            model.add(Dropout(0.2))
        model.add(Dense(units=1,activation='sigmoid',kernel_initializer='glorot_uniform'))
        model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
        return model
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
model=KerasClassifier(build_fn=create_model)
```

```python
parameters=dict(layers=[[16,8,4,2]],batch_size=[30],epochs=[100])
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
gscv=GridSearchCV(estimator=model,param_grid=parameters,cv=5)
```

```python
gscv.fit(scaled_train,y)
```

```python
gscv.cv_results_
```

```python
gscv.best_score_
```

```python
gscv.best_params_
```

```python
y_pred=gscv.predict(scaled_test)
```

```python
y_pred.shape
```

```python
y_pred=y_pred.reshape(418)
```

```python
output_df=pd.DataFrame({"PassengerId":range(892,1310),"Survived":y_pred})
```
