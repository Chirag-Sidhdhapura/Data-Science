```python
from keras.models import Sequential
from keras.layers import Dense,Dropout
```

```python
def create_model(layers):
    model=Sequential()
    for i,nodes in enumerate(layers):
        if i==0:
            model.add(Dense(units=nodes,activation='relu',kernel_initializer='he_uniform',input_dim=11))
            model.add(Dropout(0.2))
        else:
            model.add(Dense(units=nodes,activation='relu',kernel_initializer='he_uniform'))
            model.add(Dropout(0.2))
        model.add(Dense(units=1,activation='sigmoid',kernel_initializer='glorot_uniform'))
        model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
        return model
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
model=KerasClassifier(build_fn=create_model)
```

```python
parameters=dict(layers=[[22,11],[11,5]],batch_size=[10,20],epochs=[50])
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
gscv=GridSearchCV(model,parameters,cv=2)
```

```python
gscv.fit(x,y)
```
