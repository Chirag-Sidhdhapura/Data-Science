```python
sentnences=['I prefer Orange juice',
           'I prefer Orange juice',
           'I am computer Engineer',
           'Data science inquisitive',
           'Machine learning is hot topic',
           'Artificial intelligence is very tranding']
```

```python
from keras.preprocessing.text import one_hot
```

```python
one_hot_sentences=[one_hot(word,5000) for word in sentnences]
```

```python
one_hot_sentences
```

```python
from keras.preprocessing.sequence import pad_sequences
```

```python
data=pad_sequences(one_hot_sentences,maxlen=10,padding='pre')
```

```python
data
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Embedding
```

```python
model=Sequential() 
```

```python
model.add(Embedding(5000,10,input_length=10))
```

```python
model.compile('adam','mse')
```

```python
model.predict(data)
```
