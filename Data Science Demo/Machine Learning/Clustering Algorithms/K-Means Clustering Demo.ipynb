```python
import  pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("E:/tuts/codebasics Data Science/Machine Learning/kmeans.csv")
```

```python
data
```

```python
data=data.drop('Name',axis=1)
```

```python
data
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
data=StandardScaler().fit_transform(data)
```

```python
type(data)
```

```python
df=pd.DataFrame(data,columns=['Age','Income'])
```

```python
df
```

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=3)
```

```python
y_pred= model.fit_predict(df[['Age','Income']])
```

```python
y_pred
```

```python
df['y_pred']=y_pred
```

```python
df
```

```python
df1=df[df['y_pred']==0]
df2=df[df['y_pred']==1]
df3=df[df['y_pred']==2]
```

```python
plt.scatter(df1.Age,df1.Income,color='red')
plt.scatter(df2.Age,df2.Income,color='blue')
plt.scatter(df3.Age,df3.Income,color='green')
plt.xlabel("Age")
plt.ylabel("Income")
plt.legend()
plt.show() 
```

```python
r=range(1,11)
```

```python
kkkk=list()
```

```python
for k in r:
    km=KMeans(n_clusters=k)
    km.fit(df[['Age','Income']])
    kkkk.append(km.inertia_)
```

```python
len(kkkk)
```

```python
plt.plot(r,kkkk)
plt.xlim(0,10)
```
